```python
from dash import Dash, dcc, html, Input, Output
import plotly.express as px
import dash_bootstrap_components as dbc
import plotly.graph_objects as go
import plotly.figure_factory as ff
import plotly.subplots as sp
from plotly.subplots import make_subplots


app = Dash(__name__, external_stylesheets=[dbc.themes.LUX])

country_options = [{'label': country, 'value': country} for country in main['Country'].unique()]
year_options = [{'label': year, 'value': year} for year in main['Year'].unique()]

country_dropdown = dcc.Dropdown(options=country_options,clearable=False,className="m-3")
year_dropdown = dcc.Dropdown(options=year_options, clearable=True, className="m-3")

output_container = html.Div()

app.layout = dbc.Container(
    [
        html.H2("ECODash: Analysis of Economic Indicators for European Countries", style={"textAlign": "center"}, className="m-3"),
        html.Label("Select a country:"),
        country_dropdown,
        html.Label("Select a year:"),
        year_dropdown,
        dcc.Loading(output_container),
    ],
    fluid=True,
)

@app.callback(Output(output_container, "children"), [Input(country_dropdown, "value"), Input(year_dropdown, "value")])
def update(country, year):
    if country is None:
        return html.Div("Please select a country.")

    dff = main[main['Country'] == country]

    if year is not None:
        dff = dff[dff['Year'] == year]

    scatter_fig1 = go.Figure()
    scatter_fig1.add_trace(go.Scatter(
    # add trace for inflation
        x=dff['Year'],
        y=dff['Inflation'],
        mode='lines+markers+text',
        name='Inflation',
        line=dict(color='rgb(154, 191, 151)', width=2),
        marker=dict(symbol='circle'),
    ))

    scatter_fig1.add_trace(go.Scatter(
    # add trace for unemployment
        x=dff['Year'],
        y=dff['unemployment'],
        mode='lines+markers+text',
        name='Unemployment',
        line=dict(color='rgb(255, 0, 0)', width=2),
        marker=dict(symbol='circle'),
    ))

    for _,row in dff.iterrows():
        scatter_fig1.add_annotation(
        go.layout.Annotation(
            x=row["Year"],
            y=row["Inflation"],
            text=row['Inflation'],
            showarrow=False,
            align='center',
            yanchor='auto',
            yshift=10,
            textangle=-0)
       )
        scatter_fig1.add_annotation(
        go.layout.Annotation(
            x=row["Year"],
            y=row["unemployment"],
            text=row['unemployment'],
            showarrow=False,
            align='center',
            yanchor='auto',
            yshift=10,
            textangle=-0)
         )

    scatter_fig1.update_layout(
        title='Inflation and Unemployment (Annual growth rate %) in {}'.format(country),
        xaxis_title='Year',
        yaxis_title='Value',
        showlegend=True,
        template='simple_white'
    )

    scatter_fig2 = go.Figure()

    scatter_fig2.add_trace(go.Scatter(
        x=dff['Year'],
        y=dff['Imports'],
        mode='lines+markers',
        name='Imports',
        line=dict(color='rgb(154, 191, 151)', width=2),
        marker=dict(symbol='circle'),
    ))

    scatter_fig2.add_trace(go.Scatter(
        x=dff['Year'],
        y=dff['Exports'],
        mode='lines+markers',
        name='Exports',
        line=dict(color='rgb(146, 123, 139)', width=2),
        marker=dict(symbol='circle'),
    ))

    scatter_fig2.update_layout(
        title='Imports and Exports in {}'.format(country),
        xaxis_title='Year',
        yaxis_title='Imports/Exports (in millions US dollars)',
        template='simple_white'
    )


    metrics = ['Poverty', 'GNI', 'Fertility', 'Life_expectancy']
    correlation_df = main[main['Country'] == country][metrics].corr().round(2)

    heatmap = ff.create_annotated_heatmap(
        x=list(correlation_df.columns),
        y=list(correlation_df.index),
        z=correlation_df.values,
        colorscale='Blues'
    )

    heatmap.update_layout(
        title='Correlation Matrix for {}'.format(country),
        xaxis_title='Metrics',
        yaxis_title='Metrics',
        template='simple_white',
        width=910,
        height=800,
    )


    filtered_data = europe_df[europe_df['Country'] == country]
    
    rent = go.Figure()

    # Add trace for nominal Values
    rent.add_trace(go.Scatter(
        x=filtered_data[filtered_data['Indicator'] == 'Nominal']['Year'],
        y=filtered_data[filtered_data['Indicator'] == 'Nominal']['Value'],
        mode='markers',
        name='Nominal House Prices',
        marker=dict(symbol='circle', size=13),
        text='Nominal House Prices'
    ))

    # Add trace for rent Values
    rent.add_trace(go.Scatter(
        x=filtered_data[filtered_data['Indicator'] == 'Rent']['Year'],
        y=filtered_data[filtered_data['Indicator'] == 'Rent']['Value'],
        mode='markers',
        name='Real Rent Prices',
        marker=dict(symbol='square', size=13),
        text='Real Rent Prices'
    ))

    # Update layout
    rent.update_layout(
        title=f"Comparison of Nominal and Rent Prices in {country}",
        xaxis_title='Year',
        yaxis_title='Value',
        showlegend=True,
        template='simple_white',
        height=600,
    )
    rent.update_yaxes(showgrid=True)
  

    # Bar chart for total population
    population_fig = go.Figure(
        data=go.Bar(
            x=dff['Year'],
            y=dff['Population'],
            marker=dict(color='rgba(58, 123, 215, 0.6)'),
            text=dff['Population'],
        ),
    )

    population_fig.update_layout(
        title='Total Population in {}'.format(country),
        xaxis_title='Year',
        yaxis_title='Total Population',
        template='simple_white',
        height=900
    )

    fig = make_subplots(rows=1, cols=2, specs=[[{}, {}]], shared_xaxes=False,
                        shared_yaxes=False, vertical_spacing=0.001)

    fig.append_trace(go.Bar(
        x=new_df['total_savings'],
        y=new_df['Country'],
        marker=dict(
            color='rgba(50, 171, 96, 0.6)',
            line=dict(
                color='rgba(50, 171, 96, 1.0)',
                width=1),
        ),
        name='Household savings, percentage of household disposable income',
        orientation='h',
    ), 1, 1)

    fig.append_trace(go.Scatter(
        x=new_df['total_wealth'], y=new_df['Country'],
        mode='lines+markers',
        line_color='rgba(165, 187, 191 )',
        name='Household net worth, Million USD/capita',
    ), 1, 2)

    fig.update_layout(
        title='Household savings & Net worth in 2018',
        yaxis=dict(
            showgrid=False,
            showline=False,
            showticklabels=True,
            domain=[0, 0.85],
        ),
        yaxis2=dict(
            showgrid=False,
            showline=True,
            showticklabels=False,
            linecolor='rgba(102, 102, 102, 0.8)',
            linewidth=2,
            domain=[0, 0.85]
        ),
        xaxis=dict(
            zeroline=False,
            showline=False,
            showticklabels=True,
            showgrid=True,
            domain=[0, 0.42],
        ),
        xaxis2=dict(
            zeroline=False,
            showline=False,
            showticklabels=True,
            showgrid=True,
            domain=[0.47, 1],
            side='top',
            dtick=100,
        ),
        legend=dict(x=0.029, y=1.038, font_size=15),
        margin=dict(l=90, r=20, t=70, b=70),
        template='simple_white',
        width=800,
        height=700
    )

    annotations = []

    y_s = new_df['total_savings'].round(decimals=2)
    y_nw = new_df['total_wealth'].round()

    for ydn, yd, xd in zip(y_nw, y_s, new_df['Country']):
        annotations.append(dict(xref='x2', yref='y2',
                                y=xd, x=ydn - 50,
                                text='{:,}'.format(ydn) + 'M',
                                font=dict(family='sans-serif', size=14,
                                          color='rgb(128, 0, 128)'),
                                showarrow=False))
        annotations.append(dict(xref='x1', yref='y1',
                                y=xd, x=yd + 8,
                                text=str(yd) + '%',
                                font=dict(family='sans-serif', size=14,
                                          color='rgb(50, 171, 96)'),
                                showarrow=False))

    fig.update_layout(annotations=annotations)

    fig3 = make_subplots(rows=1, cols=2, specs=[[{}, {}]], shared_xaxes=False,
                        shared_yaxes=False, vertical_spacing=0.001)


    fig3.append_trace(go.Bar(
        x=new2['total_savings'],
        y=new2['Country'],
        marker=dict(
            color='rgba(50, 171, 96, 0.6)',
            line=dict(
                color='rgba(50, 171, 96, 1.0)',
                width=1),
        ),
        name='Household savings, percentage of household disposable income',
        orientation='h',
    ), 1, 1)

    fig3.append_trace(go.Scatter(
        x=new2['total_wealth'], y=new2['Country'],
        mode='lines+markers',
        line_color='rgba(165, 187, 191 )',
        name='Household net worth, Million USD/capita',
    ), 1, 2)

    fig3.update_layout(
        title='Household savings & Net worth in 2018',
        yaxis=dict(
            showgrid=False,
            showline=False,
            showticklabels=True,
            domain=[0, 0.85],
        ),
        yaxis2=dict(
            showgrid=False,
            showline=True,
            showticklabels=False,
            linecolor='rgba(102, 102, 102, 0.8)',
            linewidth=2,
            domain=[0, 0.85]
        ),
        xaxis=dict(
            zeroline=False,
            showline=False,
            showticklabels=True,
            showgrid=True,
            domain=[0, 0.42],
        ),
        xaxis2=dict(
            zeroline=False,
            showline=False,
            showticklabels=True,
            showgrid=True,
            domain=[0.47, 1],
            side='top',
            dtick=400,
        ),
        legend=dict(x=0.029, y=1.038, font_size=15),
        margin=dict(l=90, r=20, t=70, b=70),
        template='simple_white',
        width=800,
        height=700
    )

    annotations = []

    y_s = new2['total_savings'].round(decimals=2)
    y_nw = new2['total_wealth'].round()

    for ydn, yd, xd in zip(y_nw, y_s, new2['Country']):
        annotations.append(dict(xref='x2', yref='y2',
                                y=xd, x=ydn - 32,
                                text='{:,}'.format(ydn) + 'M',
                                font=dict(family='sans-serif', size=14,
                                          color='rgb(128, 0, 128)'),
                                showarrow=False))
        annotations.append(dict(xref='x1', yref='y1',
                                y=xd, x=yd + 8,
                                text=str(yd) + '%',
                                font=dict(family='sans-serif', size=14,
                                          color='rgb(50, 171, 96)'),
                                showarrow=False))

    fig3.update_layout(annotations=annotations)


    scatter_fig3 = go.Figure()

    scatter_fig3.add_trace(go.Scatter(
            x=dff['Year'],
            y=dff['Women_unemp'],
            mode='lines+markers+text',
            marker=dict(
                color='rgba(255, 0, 0, 0.6)',
                line=dict(
                    color='rgba(255, 0, 0, 1.0)',
                    width=1),
            ),
            name='Native Women: Unemployment',
    ))
    
    scatter_fig3.add_trace(go.Scatter(
         x=dff['Year'],
         y=dff['Men_unemp'],
         mode='lines+markers+text',
         marker=dict(
                 color='rgba(0, 0, 222, 0.6)',
                 line=dict(
                 color='rgba(0, 0, 222, 1.0)',
                width=1),
        ),
        name='Native Men: Unemployment',
    
    ))

    scatter_fig3.update_layout(
        title='Women and Men Native Unemployment in {}'.format(country),
        xaxis_title='Year',
        yaxis_title='Unemployment by Gender',
        template='simple_white'
    )

    for _,row in dff.iterrows():
        scatter_fig3.add_annotation(
        go.layout.Annotation(
            x=row["Year"],
            y=row["Women_unemp"],
            text=row['Women_unemp'],
            showarrow=False,
            align='center',
            yanchor='auto',
            yshift=10,
            textangle=-0)
       )
        scatter_fig3.add_annotation(
        go.layout.Annotation(
            x=row["Year"],
            y=row["Men_unemp"],
            text=row['Men_unemp'],
            showarrow=False,
            align='center',
            yanchor='auto',
            yshift=10,
            textangle=-0)
         )
        
        fig4 = go.Figure()

        fig4 = px.line(main, x='Year', y='unemployment', color='Country', title='Unemployment by Country', line_dash='Country',markers=True, line_shape='linear')
        fig4.update_layout(template='simple_white', width=1500, height=700)


    return html.Div(
        [
            dbc.Row(
                [
                    dbc.Col(dcc.Graph(figure=scatter_fig1, className="border"), lg=6),
                    dbc.Col(dcc.Graph(figure=scatter_fig2, className="border"), lg=6),
                ],
                className="mt-4",
            ),
            dbc.Row(
                [
                    dbc.Col(dcc.Graph(figure=rent, className="border"), lg=6),
                    dbc.Col(dcc.Graph(figure=heatmap, className="border"), lg=6),
                ],
                className="mt-4",
            ),
            dbc.Row(
                dbc.Col(dcc.Graph(figure=population_fig, className="border"), lg=12),
                className="mt-4",
            ),
            dbc.Row(
                [
                dbc.Col(dcc.Graph(figure=fig, className="border"), lg=6),
                dbc.Col(dcc.Graph(figure=fig3, className="border"), lg=6),
                dbc.Col(dcc.Graph(figure=scatter_fig3, className="border"), lg=12),
            ],
            className="mt-4",
            ),
            dbc.Row(
               [
                dbc.Col(dcc.Graph(figure=fig4, className="border"), lg=12),
            ],
            className="mt-4",
        ),

        ]
    )       
if __name__ == "__main__":
    app.run_server(port=8080)
```
